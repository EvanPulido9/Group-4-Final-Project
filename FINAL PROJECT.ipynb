```python
no_of_matches = df['season'].value_counts().sort_index()
no_of_matches
```

```python
no_of_matches.plot(kind='bar' ,color='blue')
plt.title('Number of Matches Played per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches Played')
plt.show()
```

```python
toss_won = df['toss_won'].value_counts().sort_values(ascending=True)
toss_won
```

```python
toss_won.plot(kind='barh',color='#2a5233')
plt.title('Number of Tosses won by each Team')
plt.xlabel('Number of Tosses Won')
plt.ylabel('Team Name')
plt.show()
```

```python
bowl_first = df['decision'].value_counts()[0]
bat_first = df['decision'].value_counts()[1]
```

```python
plt.pie([bowl_first,bat_first],labels=['Bowl First','Bat First'],autopct='%1.1f%%')
plt.title('Bat First VS Bowl First Ratio')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(y='season',x='home_captain',data=df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='venue_name')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(7,7))
sns.stripplot(x='home_runs',y='venue_name',data=df,hue='home_team')
plt.legend(loc=(1,0.1))
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(y='home_boundaries',x='home_team',data=df,hue='home_team')
plt.xticks(rotation=45)
plt.legend(loc=(1,.1))
```

```python
sns.violinplot(y='season',x='super_over',data=df)
```

```python
df['year'] = pd.to_datetime(df['start_date']).dt.year

season_avg_scores = df.groupby('year')['home_runs'].mean()

plt.figure(figsize=(8, 6))
plt.plot(season_avg_scores.index, season_avg_scores.values, marker='o', color='orange')
plt.title('Seasonal Trend Analysis')
plt.xlabel('Year')
plt.ylabel('Average Scores')
plt.xticks(rotation=90)
plt.show()
```

```python
pom_frequency = df['pom'].value_counts().head(20)

plt.figure(figsize=(10, 6))
plt.barh(pom_frequency.index, pom_frequency.values, color='purple')
plt.title('Player of the Match Analysis')
plt.xlabel('Frequency')
plt.ylabel('Player')
plt.show()
```

```python
season_win_percentage = df.groupby('season')['winner'].apply(lambda x: x.value_counts(normalize=True)).unstack()

plt.figure(figsize=(12, 6))
for team in season_win_percentage.columns:
    plt.plot(season_win_percentage.index, season_win_percentage[team], marker='o', label=team)
plt.title('Season-wise Win Percentage')
plt.xlabel('Season')
plt.ylabel('Win Percentage')
plt.legend()
plt.show()
```
