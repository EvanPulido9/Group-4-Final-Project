```python
season_win_percentage = df.groupby('season')['winner'].apply(lambda x: x.value_counts(normalize=True)).unstack()

plt.figure(figsize=(12, 6))
for team in season_win_percentage.columns:
    plt.plot(season_win_percentage.index, season_win_percentage[team], marker='o', label=team)
plt.title('Season-wise Win Percentage')
plt.xlabel('Season')
plt.ylabel('Win Percentage')
plt.legend()
plt.show()
```

```python
df = df.dropna()
df_model = df.filter([
    'home_team','away_team', 'toss_won', 'decision', 'winner',
    'home_captain', 'away_captain', 'pom',
    'home_overs', 'home_wickets', 'home_boundaries',
    'away_overs', 'away_wickets', 'away_boundaries'
])
```

```python
df_model.head()
```

```python
home_team_wins = []

for i in df_model.index:
    if df_model['home_team'][i] == df_model['winner'][i]:
        home_team_wins.append(1)
    else:
        home_team_wins.append(0)
print(home_team_wins)
```

```python
df_model["home_team_winner"] = home_team_wins
df_model.head()
```

```python
home_toss_wins = []

for i in df_model.index:
    if df_model['home_team'][i] == df_model['toss_won'][i]:
        home_toss_wins.append(1)
    else:
        home_toss_wins.append(0)
print(home_toss_wins)
```

```python
df_model['home_wins_toss'] = home_toss_wins
df_model.head()
```

```python
df_model2 = df_model.drop("home_team", axis=1)
df_model2 = df_model2.drop("away_team", axis=1)
df_model2 = df_model2.drop("toss_won", axis=1)
df_model2 = df_model2.drop("winner", axis=1)
df_model2 = df_model2.drop("decision", axis=1)
df_model2 = df_model2.drop("home_captain", axis=1)
df_model2 = df_model2.drop("away_captain", axis=1)
df_model2 = df_model2.drop("pom", axis=1)
df_model2.head()
```

```python
X = df_model2.drop('home_team_winner', axis=1)
y = df_model2['home_team_winner']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
log_regression = LogisticRegression()

#fit the model using the training data
log_regression.fit(X_train,y_train)

#use model to make predictions on test data
y_pred = log_regression.predict(X_test)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("F1 score:",metrics.f1_score(y_test, y_pred))
```

```python
for i in log_regression.coef_.index:
    print(X[i], log_regression.coef_[i])
```
